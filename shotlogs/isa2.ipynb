```python
import pandas as pd
print(pd.__version__)
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('data/shot_logs.csv')
all_players = sorted(pd.unique(df[['player_name']].values.ravel()))
```

# Top 5 Shooters 3 puntlijn

```python
array = []
def top_3puntlijn(playername):
    df3 = df[(df['player_name'] == str(playername)) & (df['SHOT_DIST'] >= 22.0)]
    obj = df3.groupby(['player_name','SHOT_RESULT']).size()
    print(obj)
    dict = {}
    try:
        dict = {
            'player' : playername,
            'made_shots': obj[0]
           }
    except IndexError:
        dict = {
            'player': playername,
            'made_shots' : 0
        }
    return dict

for player in all_players:
    dict = top_3puntlijn(player)
    array.append(dict)

```

```python
print(array)
```

```python
array.sort(key=lambda x: x['made_shots'], reverse=True)
```

```python
puntlijn_players = array[:5]
```

```python
print(puntlijn_players)
```

```python
puntlijnplayer = []
puntlijnshots = []
for p in puntlijn_players:
    puntlijnplayer.append(p['player'])
    puntlijnshots.append(p['made_shots'])
    
print (puntlijnplayer)
```

## Percentage

```python
def top_3puntlijnpercent(playername):
    df3 = df[(df['player_name'] == str(playername)) & (df['SHOT_DIST'] >= 22.0)]
    obj = df3.groupby(['player_name', 'SHOT_RESULT']).size()
    print(obj)
    dict = {}
    try: 
        dict = {
            'player' : playername,
            'made_shots': obj[0],
            'missed_shots': obj[1]
        }
    except IndexError: 
        dict = {
            'player' : playername,
            'made_shots': 0,
            'missed_shots': 0
        }
    return dict

array_percent = []
for player in all_players:
    dict = top_3puntlijnpercent(player)
    array_percent.append(dict)
```

```python
percentageliner = []
for a in array_percent:
    try:
        percentageliner.append({'player': a['player'], 'shotpercentage': (a['made_shots']/(a['made_shots'] + a['missed_shots']))*100})
        #percentageliner.append((a['made_shots']/(a['made_shots'] + a['missed_shots']))*100)
    except ZeroDivisionError:
        percentageliner.append({'player': a['player'], 'shotpercentage': 0})

        
#print(percentageliner)
        
percentageliner.sort(key=lambda x: x['shotpercentage'], reverse=True)

percentageliner = percentageliner[:5]
```

```python
puntlijnplayerpercentage = []
puntlijnshotspercentage = []
puntperc = []
for p in percentageliner:
    puntlijnplayerpercentage.append(p['player'])
    puntlijnshotspercentage.append(p['shotpercentage'])
    
puntlijnshotspercentage = [ '%.2f' % elem for elem in puntlijnshotspercentage ]
for p in puntlijnshotspercentage:
    puntperc.append(float(p))

print (puntlijnplayerpercentage)
print (puntperc)
```
